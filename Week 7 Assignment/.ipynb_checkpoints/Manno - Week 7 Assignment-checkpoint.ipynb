```python
import numpy as np
import pandas as pd
#set some pandas options controling output format
pd.set_option('display.notebook_repr_html',True) # output as flat text and not HTML
pd.set_option('display.max_rows', None) # this is the maximum number of rows we will display
pd.set_option('display.max_columns', None) # this is the maximum number of rows we will display

# bring in matplotlib for graphics
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 1. Load the ratings by user information that you collected into a pandas dataframe.
# read the CSV into a dataframe
ratings = pd.read_csv('data/is362Flix.csv', index_col=0)
ratings.head()  # take a peak at the date.
```

```python
# 2.a. Show the average ratings for each USER
# Calculate the average by column
ratings.mean()
```

```python
# 2.b. Show the average ratings for each MOVIE
# Calculate the average by row 
rmean = pd.Series(ratings.mean(axis=1))
rmean
```

```python
# 3.a. Create a new pandas dataframe, with normalized ratings for each user. Again, show
#      the average ratings for each user and each movie.
# To calculate this I will be using the following formula mentioned in this post:
#      http://bi-analytics.org/topic/9-standardization-vs-normalization/
# The formula is as follows:
#      Xnew = (X-Xmin) / (Xmax - Xmin
# Calculate the normalized ratings for each cell. And transpose the rows and columns
ratings_norm = ((ratings - ratings.min()) / (ratings.max() - ratings.min())).transpose()
ratings_norm    # Display the dataframe
```

```python
# 3.b. Show the average NORMALIZED ratings for each USER
# Calculate the average of the normalized ratings by row.
ratings_norm.mean(axis=1)
```

```python
# 3.c. Show the average NORMALIZED ratings for each MOVIE
# Calculate the average of the normalized ratings by column.
rnorm = pd.Series(ratings_norm.mean())
rnorm
```

```python
# create a dataframe with both the average and normalized series by movie.
ratings_averages = pd.DataFrame({'mean':rmean, 'norm':rnorm},rmean.index)
ratings_averages
```

```python
# Sort the movies by the AVERAGE rating
ratings_averages.sort_values('mean',ascending=False)
```

```python
# Sort the movies by the NORMALIZED average rating
ratings_averages.sort_values('norm',ascending=False)
```

## Conclusions
I can see what normalizing the data is doing. It's trying to control for the harshness of a critic. For instance if a user has a policy of never giving a movie a 5 out of 5, then realy the top rating a movie can recieve would be a 4. Normalizing the data attempts to remove the skew of each user's proclivaties.

Normalization also appears be an to attempt to create a scale for each user. Meaning rating all movies base on where they fall between the users worst rated movie and the users highest rated movie. 

That being said looking at the standings of average rating vs normalzed rating there is one interesting change betweent the data sets:
### Legos vs. La La Land
in the average dataset *The Lego Movie* is 5th and *La La Land* is 7th. In the normalized average these positions are flipped. *La La Land* is 5th and *The Lego Movie* is 7th. 

The interesting thing about this flip is that judging from the ratings *The Lego Movie* appears to be the better choice for most people. Below are the datasets both movies.

The_Lego_Movie:	3, 3, 4, 3, 4, 3, 3, 3, 4, 5

La_La_Land:		4, 1, 4, 4, 2

As you can see *The Lego Movie's* lowest rating is a 3 while *La La Land's* lowest rating is a 1. It appears that *The Lego Movie* is "generally" liked, meaning no one really hates it. However, *La La Land* appears to be loved or hated. If I were to reccomend a movie to a stranger I would be more inclined to reccomend *The Lego Movie* as it would appear at worst they wont like it too much. On the other hand, a stranger might hate *La La Land*.

There is one more feature of normalization that is worth mentioning. In this assignment it's not really relavent, but normalization creates a number that can make two different review scales comperable. The scale for this assignment was to gather ratings between 1 and 5; however, if I gathered ratings from a source with a scale between 1 and 10 I could use normalization to easilly compare movie ratings on the different scales.

```python
# 5.a. [Extra credit] Create another new pandas dataframe, with STANDARDIZED ratings for each user.
#      Once again, show the average ratings for each user and each movie.
# To calculate this I will be using the following formula mentioned in this post:
#      http://bi-analytics.org/topic/9-standardization-vs-normalization/
# The formula is as follows:
#      Xnew = (X-μ) / σ
# Calculate the normalized ratings for each cell. And transpose the rows and columns
ratings_stand = ((ratings - ratings.mean()) / (ratings.std())).transpose()
ratings_stand
```

```python
# 5.b. Show the average NORMALIZED ratings for each USER
# Calculate the average of the normalized ratings by row.
ratings_stand.mean(axis=1)
```

```python
# 5.C. Show the average NORMALIZED ratings for each MOVIE
# Calculate the average of the normalized ratings by column.
rstand = pd.Series(ratings_stand.mean())
rstand
```

```python
# Add the average standardized rating for each movie to the dataframe that already 
# contains the average and normamalized average movie rating.
ratings_averages['stand'] = rstand
ratings_averages
```

```python
# Sort the movies by the AVERAGE rating
ratings_averages.sort_values('mean',ascending=False)
```

```python
# Sort the movies by the NORMALIZED average rating
ratings_averages.sort_values('norm',ascending=False)
```

```python
# Sort the movies by the STANDARDIZED average rating
ratings_averages.sort_values('stand',ascending=False)
```

## Conclusions
A standardized value tells us how may standard deviations away from the mean an rating is. I've standardized the values based on each user. As with normalization, stnadardization has the effect of allowing ratings on different scales to be compared to each other. In reviewing the results the results returned from the standardized results are close to the average results, with one exception Central Intelligence and Chips positions are flipped. This may be attributable to the limited number of datapoints for the movie Chips. As more ratings for Chips get added to the dataset it will more accurately reflect where chips should fall in the rankings.
